# Singular/Plural Object Types

```python
import glob
import pandas as pd

objects_df = pd.read_hdf('/home/richard/Git/bbk-mqaf/data/latest/field.h5')
```

```python
obj_types = objects_df['[objectType'].value_counts()
total_objects = len(objects_df)
```

```python
# Now extract all those ending with 's' - not comprehensive plural detection system of course

plural_types = obj_types[obj_types.index.str.endswith("s")]
```

```python
# Remove those ending with 'ss' or 'es' to get rid of some obvious false positives

plural_types = plural_types[~plural_types.index.str.endswith("es")]
plural_types = plural_types[~plural_types.index.str.endswith("ss")]
plural_types = plural_types[~plural_types.index.str.endswith("us")]
```

```python
# Now go through each, remove the 's', and check if it also exists in original list (upper case first letter for the comparision) 

# Index - Singular Name, Count, Plural name, count
duplicates = pd.DataFrame()
total_plural = 0

data = []
for index, value in plural_types.items():
    singular = index[:-1].capitalize()
    if singular in obj_types:
        count = obj_types.get(singular)
        data.append([singular, count, index, value])
        total_plural += value

duplicates = pd.DataFrame(data, columns = ["Singular Name", "Singular Objects", "Plural Name", "Plural Count"])
```

```python
print(f"Total Objects {total_objects} of which {total_plural} ({(total_plural/total_objects)*100:.4f}%) are possible plural duplicates")
```

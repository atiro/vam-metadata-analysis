```python
# Field Exisistance - Per department, Show all fields in tables (grouped by category) with existance counts/ percentage
```

```python

import glob
import pandas as pd

df = pd.concat(map(pd.read_csv, glob.glob('/home/richard/Git/metadata/vam-metadata/data/*.csv')))
df
```

```python
complete_df = df.filter(regex='collectionCode|existence*', axis=1)
# Add object column so we have a count
complete_df['object'] = 1
```

```python
# Go through each department in turn. Start with PDP

grouped_df = complete_df.groupby([' collectionCode']).sum()

```

```python
grouped_df
```

```python

percentage_df = grouped_df.iloc[:, 1:-1].divide(grouped_df.iloc[:,-1], axis = 'rows')
percentage_df
```

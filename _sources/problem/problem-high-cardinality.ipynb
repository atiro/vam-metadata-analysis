```python

import glob
import pandas as pd

objects_df = pd.read_hdf('/home/richard/Git/bbk-mqaf/data/latest/field.h5')
```

```python
complete_df = objects_df.filter(regex='collectionCode|cardinality*', axis=1)
complete_df
```

```python
# Filter out anything less than 2 per field
#complet_df = complete_df[" cardinality:artistMakerPerson"].value_counts().reset_index()
```

```python
 
import altair as alt

vcharts = []
hcharts = []
charts = []
count =0

  # Now need to loop over each cardinality  in turn
for column in complete_df.columns[1:]:
    
    dept_col_counts = complete_df[column].value_counts().reset_index()
    dept_col_df = pd.DataFrame(dept_col_counts).iloc[1:]
#    print(dept_col_df)
    if(len(dept_col_df) < 5):
        continue
    
    chart = alt.Chart(dept_col_df).mark_bar().encode(
      x=alt.X('index:O', axis=alt.Axis(title='Cardinality')),
      y=alt.Y(column, type='quantitative',  axis=alt.Axis(title='Objects'))
    ).properties(title=column.split(":")[1])
    charts.append(chart)
    if count > 1:
      hcharts.append(alt.hconcat(*charts))
      charts = []
      count = 0
    else:
      count += 1

alt.vconcat(*hcharts)
```

# Factcheck (Techniques)

This compares the object's production dates with the object' techniques, and identifies any cases where a technique is stated that did not exist at the time of the object's known production. This could indicate:
  * A typo in dates
  * A production date range given that could be more tightly specified (e.g. if a technique was only available from a certain date, this sets the earliest bound of the date range, rather than a century or decade range being used)


```python
import glob
import pandas as pd

objects_df = pd.read_hdf('/home/richard/Git/bbk-mqaf/data/latest/factcheck-techniques.h5')
```

```python
# Remove objects with positive score

falseobjects_df = objects_df[objects_df[" ruleCatalog:score]"] < 0]
```

```python
print(f"Total Objects {len(objects_df)} of which {len(falseobjects_df)} contain possible dating or material errors")
```

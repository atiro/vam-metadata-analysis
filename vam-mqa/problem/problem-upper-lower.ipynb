# Upper/Lower Object Type

```python

import glob
import pandas as pd

objects_df = pd.read_hdf('/home/richard/Git/bbk-mqaf/data/latest/field.h5')
```

```python
total_objects = len(objects_df)
```

```python
obj_types = objects_df['[objectType'].value_counts()
```

```python
lower_types = obj_types[obj_types.index.str.islower()]
```

## Upper/Lower Case Object Types

```python
# Now go through each with lowercase, and check if it also exists in original list (upper case first letter for the comparision) 

# Index - Singular Name, Count, Plural name, count
duplicates = pd.DataFrame()
total_lower = 0
total_noupper = 0

data = []
data_noupper = []
for index, value in lower_types.items():
    upper = index.capitalize()
    if upper in obj_types:
        count = obj_types.get(upper)
        total_lower += value
        data.append([upper, count, index, value])
    else:
        # Even stranger - no uppercase version exists - probably a typo
        total_noupper += value
        data_noupper.append([index, value])

duplicates = pd.DataFrame(data, columns = ["Uppercase Name", "Uppercase Objects", "lowercase Name", "lowercase Count"])
     
noupper = pd.DataFrame(data_noupper, columns = ["lowercase Name", "lowercase Count"])   

duplicates.sort_values(by=['lowercase Count'], ascending=False).head(25)
```

```python
print(f"Total Objects {total_objects} of which {total_lower} ({(total_lower/total_objects)*100:.4f}%) are possible plural duplicates")
```

```python
duplicates.sort_values(by=['lowercase Count'], ascending=False).head(25)
```

```python
print(f"Total Objects {total_objects} of which {total_noupper} ({(total_noupper/total_objects)*100:.4f}%) are lower case only")
```

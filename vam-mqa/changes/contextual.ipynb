```python
# Graph one - BY collection code show contextual completeness
```

```python
# First create new data frame reducing down to just the columns we need

complete_df = df.filter(regex='collectionCode|completeness*', axis=1)
complete_df
```

```python
# Add a column to keep track of how many objects per department

complete_df['object'] = 1
complete_df
```

```python
# Then reduce down to sums aggregating on collectionCode

grouped_df = complete_df.groupby(['[collectionCode']).sum()
grouped_df
```

```python
# Then convert to percentages

percentage_df = grouped_df.iloc[:, 1:-1].divide(grouped_df.iloc[:,-1], axis = 'rows')
percentage_df
```

```python
# Tidy up column names
import re

percentage_df = percentage_df.rename(columns=lambda x: re.sub('completeness:','',x))
percentage_df
```

```python
import altair as alt

vcharts = []
hcharts = []
charts = []
dept = None
indexn = None
count = 0
for index, row in percentage_df.iterrows():
    dept = row.to_frame(name='concepts')
    charts.append(
        alt.Chart(dept.reset_index()).mark_bar().encode(
        x=alt.X('index:N', axis=alt.Axis(title='Cataloguing Area')),
        y=alt.Y('concepts',  axis=alt.Axis(format='%', title='Completeness'))
       ).properties(title=index))
    if count > 3:
      hcharts.append(alt.hconcat(*charts))
      charts = []
      count = 0
    else:
      count += 1

alt.vconcat(*hcharts)
```

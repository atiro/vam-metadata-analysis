```python
# Visualising Completeness and Cardinality
```

```python
import pandas as pd

df_objects = pd.read_csv("objects-rules-150000-group.csv", index_col=False)
df_objects.head(20)
```

```python
pd.set_option('display.max_rows', 50)
df_objects.sum()
```

```python

df_existence = df_objects[df_objects['completeness:TOTAL'].between(0.45, 0.46)]
df_existence = df_existence.filter(regex='existence')
df_existence.sum()
```

```python
# Bin overall record completeness scores into percentages

completeness = pd.cut(df_objects['completeness:TOTAL'], bins=[x * 0.01 for x in range(0, 100)],labels=range(0,99)).value_counts()
completeness.reset_index().rename(columns={0:'counts'})
```

```python
# Graph percentages
alt.Chart(completeness.reset_index().rename(columns={0:'counts'})).mark_bar().encode(
    x='index:O',
    y='completeness:TOTAL:Q'
)
```

```python
df_existence = df_objects[df_objects['completeness:TOTAL'].between(0.72, 0.73)]
df_existence = df_existence.filter(regex='existence')
existence_019 = pd.DataFrame(df_existence.sum()).reset_index().rename(columns={'index':'field', 0:'count'})
```

```python
len(df_existence.index)
```

```python
num_rows = len(df_existence)
existence_019_exists = existence_019[existence_019['count'].between(1, num_rows-1)]
existence_019_exists['count'] = existence_019_exists['count'] / num_rows
existence_019_exists.head(20)
```

```python
bars = alt.Chart(existence_019_exists).mark_bar().encode(
    x='count:Q',
    y=alt.Y("field:O", sort='-x')
)

text = bars.mark_text(
    align='left',
    baseline='middle',
    dx=3  # Nudges text to right so it doesn't appear on top of the bar
).encode(
    text=alt.Text('count:Q', format=',.2r')
)

(bars + text).properties(height=500)
```

```python
df_concepts = df_objects[df_objects['completeness:TOTAL'].between(0.60, 0.61)]
df_concepts = df_concepts.filter(regex='completeness')
df_concepts = df_concepts.mask(df_concepts > 0, 1)
concepts_60 = pd.DataFrame(df_concepts.sum()).reset_index().rename(columns={'index':'field', 0:'count'})
```

```python
df_concepts.head()
```

```python
concepts_60.head()
```

```python
num_rows = len(df_concepts)
concepts_60_exists = concepts_60[concepts_60['count'].between(1, num_rows)]
concepts_60_exists['count'] = concepts_60_exists['count'] / num_rows
concepts_60_exists.head(20)
```

```python
bars = alt.Chart(concepts_60_exists).mark_bar().encode(
    x='count:Q',
    y=alt.Y("field:O", sort='-x')
)

text = bars.mark_text(
    align='left',
    baseline='middle',
    dx=3  # Nudges text to right so it doesn't appear on top of the bar
).encode(
    text=alt.Text('count:Q', format=',.2r')
)

(bars + text).properties(height=500)
```
